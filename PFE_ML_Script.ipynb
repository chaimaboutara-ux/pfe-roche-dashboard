*Targeted Data Augmentation Script*

```python
import pandas as pd
import random

# 1. Load your original clean dataset
df_core = pd.read_csv('Insights_Final_Cleaned_100.csv')

# 2. Define the Target Columns (your multi-label categories)
target_cols = [
    'Is_Actionable', 'Cat_Access_And_Policy', 'Cat_Competitor_Intelligence',
    'Cat_Medical_Education_&_Practice', 'Cat_Operations_And_Diagnostics',
    'Cat_Stakeholder_Mapping', 'Cat_Strategic_Partnership', 'Cat_Promotional_Activity'
]

# Print current data balance
print("--- CURRENT CATEGORY COUNTS ---")
for col in target_cols:
    print(f"{col}: {df_core[col].sum()} rows")

# 3. Dictionaries of Tunisian Pharmaceutical Vocabulary for randomized cloning
products = ['Hemlibra', 'Vabysmo', 'Phesgo', 'Ocrevus', 'Tecentriq', 'Kadcyla', 'Perjeta', 'Alecensaro']
territories = ['Hematology', 'Ophthalmology', 'Oncology', 'Neurology']
cities = ['Tunis', 'Sfax', 'Sousse', 'Jendouba', 'Gafsa', 'Monastir', 'Bizerte', 'Gabes']
hospitals = ['Hopital Aziza Othmana', 'Hopital Charles Nicolle', 'Hopital Habib Bourguiba de Sfax',
             'CHU Farhat Hached de Sousse', 'Military Hospital of Tunis', 'Institut Hédi Raïs']
kols = ['Pr. Gouider', 'Pr. Chebil', 'Dr. Aloulou', 'Dr. Khanfir', 'Pr. Triki', 'Dr. Bouraoui']
competitors = ['Bayer (Eylea)', 'Novartis', 'Biosimilar competitor', 'Local generic alternative']

# 4. Define rich templates for your under-represented categories
templates = {
    'Cat_Competitor_Intelligence': [
        {"title": "Competitor Activity: {comp} Market Share",
         "desc": "During the local meeting in {city}, {kol} highlighted that {comp} is offering aggressive discount packs to private clinics to secure their retina patient flow before our product gains traction."},
        {"title": "{comp} Symposium announced",
         "desc": "An upcoming regional congress will feature a dedicated symposium hosted by {comp}. They aim to present new real-world data to challenge our positioning in {territory}."},
        {"title": "Competitor switch at {hosp}",
         "desc": "{kol} mentioned that due to budget limits at {hosp}, administrative procurement is leaning towards a {comp} contract next semester."}
    ],
    'Cat_Stakeholder_Mapping': [
        {"title": "New Committee election at STO",
         "desc": "{kol} is expected to lead the new strategic board for {territory} in {city}, which will heavily influence institutional protocols next year."},
        {"title": "Key decision maker mapping at {hosp}",
         "desc": "We mapped the internal decision flow at {hosp}. The head of the pharmacy department works closely with {kol} for validating innovative drug requests."},
        {"title": "HCP Network expansion in the South",
         "desc": "New generation of young ophthalmologists in {city} are looking up to {kol} for clinical guidance. Strategic mapping of these fast-movers is required."}
    ],
    'Cat_Strategic_Partnership': [
        {"title": "MOH Screening Campaign Collaboration",
         "desc": "Discussion initiated with {kol} regarding a national diagnostic screening framework supported by {hosp} to minimize under-diagnosis in {territory}."},
        {"title": "Institutional donation follow-up",
         "desc": "The logistics for the equipment donation at {hosp} in {city} are being finalized. This partnership will optimize the patient journey significantly."},
        {"title": "Scientific Society Joint Workshop",
         "desc": "Roche is partnering with the national society to sponsor an advanced preceptorship program co-facilitated by {kol}."}
    ],
    'Cat_Operations_And_Diagnostics': [
        {"title": "Injection Room bottlenecks at {hosp}",
         "desc": "{kol} reported a severe delay in patient flow due to the lack of dedicated OCT devices and injection rooms in {city}. Patients are waiting up to 3 weeks."},
        {"title": "Under-diagnosis challenge in {territory}",
         "desc": "Statistical update from {city} shows nearly 45% of mild cases remain undiagnosed because local centers lack specialized screening kits."},
        {"title": "Patient Flow Optimization Project",
         "desc": "An audit of the day-hospital capacity at {hosp} suggests that digital tracking tools could double the diagnostic speed for {territory}."}
    ]
}

# 5. Automatic Equitable Augmentation Loop
# We want EVERY weak category to have at least 45 examples to match the strong categories.
TARGET_MINIMUM = 45
mock_rows_list = []

for category, sample_templates in templates.items():
    current_count = df_core[category].sum()

    if current_count < TARGET_MINIMUM:
        rows_needed = TARGET_MINIMUM - current_count
        print(f"\n🚀 Category [{category}] has only {current_count} rows. Generating {rows_needed} clones...")

        for _ in range(rows_needed):
            # Pick a random template sentence for this specific category
            chosen_template = random.choice(sample_templates)

            # Populate variables randomly to create unique phrasing
            prod = random.choice(products)
            city = random.choice(cities)
            hosp = random.choice(hospitals)
            kol = random.choice(kols)
            comp = random.choice(competitors)
            terr = random.choice(territories)

            formatted_title = chosen_template["title"].format(prod=prod, city=city, hosp=hosp, kol=kol, comp=comp, territory=terr)
            formatted_desc = chosen_template["desc"].format(prod=prod, city=city, hosp=hosp, kol=kol, comp=comp, territory=terr)

            # Create a mock data record matching the old layout structure
            new_row = {
                'insight id': f"MOCK_{random.randint(100000, 999999)}",
                'Date': '2026-05-18',
                'insight title': formatted_title,
                'insight description': formatted_desc,
                'Category_ML': category.replace('Cat_', '').replace('_', ' '),
                'Product_Clean': prod,
                'hcp': kol,
                'hco': hosp,
                'country': 'Tunisia',
                'Territory': terr,
                'created by': 'AI Augmentor Script',
                'voting': 0,
                # Clear all labels to 0 first, then set the correct multi-label targets
                'Is_Actionable': random.choice([0, 1]), # Actionability occurs naturally
                'Cat_Access_And_Policy': 0,
                'Cat_Competitor_Intelligence': 1 if category == 'Cat_Competitor_Intelligence' else 0,
                'Cat_Medical_Education_&_Practice': 0,
                'Cat_Operations_And_Diagnostics': 1 if category == 'Cat_Operations_And_Diagnostics' else 0,
                'Cat_Stakeholder_Mapping': 1 if category == 'Cat_Stakeholder_Mapping' else 0,
                'Cat_Strategic_Partnership': 1 if category == 'Cat_Strategic_Partnership' else 0,
                'Cat_Promotional_Activity': 0
            }
            mock_rows_list.append(new_row)

# 6. Merge real and mock data together
df_mock = pd.DataFrame(mock_rows_list)
df_augmented_final = pd.concat([df_core, df_mock], ignore_index=True)

# Save the newly equitable database
df_augmented_final.to_csv('Insights_Master_Augmented.csv', index=False)

print("\n" + "="*40)
print(f"✅ SUCCESS: Dataset expanded from {len(df_core)} to {len(df_augmented_final)} rows!")
print("="*40)
print("--- NEW BALANCED CATEGORY COUNTS ---")
for col in target_cols:
    print(f"{col}: {df_augmented_final[col].sum()} rows")
```

**Training the Model**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# ==========================================
# 1. LOAD DATA & SPLIT
# ==========================================
print("⏳ Loading dataset and generating dynamic splits...")
# Load the augmented dataset created by your clone script
df_master = pd.read_csv('Insights_Master_Augmented.csv')

# Dynamic 80% Training and 20% Testing split in memory
train_df, test_df = train_test_split(df_master, test_size=0.20, random_state=42)

print(f"✅ Master Set size: {len(df_master)} rows")
print(f"✅ Training Set: {len(train_df)} rows | Testing Set: {len(test_df)} rows\n")

# ==========================================
# 2. FEATURE ENGINEERING & TEXT VECTORIZATION
# ==========================================
print("⏳ Injecting categorical context and vectorizing text...")

# Combine context columns so the text features contain product and territory hints
X_train_text = (
    train_df['Product_Clean'].fillna('General') + " " +
    train_df['Territory'].fillna('General') + " " +
    train_df['insight title'].fillna('') + " " +
    train_df['insight description'].fillna('')
)

X_test_text = (
    test_df['Product_Clean'].fillna('General') + " " +
    test_df['Territory'].fillna('General') + " " +
    test_df['insight title'].fillna('') + " " +
    test_df['insight description'].fillna('')
)

# Advanced Vectorizer looking at words and phrases up to 3 words long
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 3), stop_words='english')

X_train = vectorizer.fit_transform(X_train_text)
X_test = vectorizer.transform(X_test_text)

# Define the target binary categories
target_cols = [
    'Is_Actionable', 'Cat_Access_And_Policy', 'Cat_Competitor_Intelligence',
    'Cat_Medical_Education_&_Practice', 'Cat_Operations_And_Diagnostics',
    'Cat_Stakeholder_Mapping', 'Cat_Strategic_Partnership', 'Cat_Promotional_Activity'
]

y_train = train_df[target_cols]
y_test = test_df[target_cols]

# ==========================================
# 3. MODEL INITIALIZATION & TRAINING (SAFE MODE)
# ==========================================
print("⏳ Training Random Forest Classifier on all target categories...")

# n_jobs=1 prevents multi-core processor deadlocks and runs safely instantly
forest_estimator = RandomForestClassifier(n_estimators=150, random_state=42)
model = MultiOutputClassifier(forest_estimator, n_jobs=1)

# Fit the algorithm
model.fit(X_train, y_train)
print("✅ Model trained successfully in safe-execution mode!\n")

# ==========================================
# 4. PREDICTION AND EVALUATION
# ==========================================
print("⏳ Generating predictions and analyzing evaluation metrics...")
y_pred = model.predict(X_test)

# Print a classification report for each strategic pillar
print("=====================================================================")
print("🎯 MULTI-LABEL AFFILIATE PERFORMANCE REPORT")
print("=====================================================================")

for idx, col in enumerate(target_cols):
    print(f"\n📊 Pillar Category: [ {col} ]")
    # Extract prediction vector vs true testing vector for this column
    true_labels = y_test[col]
    predicted_labels = y_pred[:, idx]

    print(classification_report(true_labels, predicted_labels, zero_division=0))
    print("-" * 65)

# Calculate global exact match subset accuracy
subset_acc = accuracy_score(y_test, y_pred)
print(f"\n⭐ Overall Exact Match Ratio (Subset Accuracy): {subset_acc * 100:.2f}%")
print("=====================================================================")
```

**Create the Dashboard Code (app.py)**

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# 1. Page Configuration (The tab title and layout)
st.set_page_config(page_title="Roche Tunisia Field Insights Dashboard", layout="wide")

# 2. Load the augmented dataset
@st.cache_data # This keeps the dashboard running incredibly fast
def load_data():
    df = pd.read_csv('Insights_Master_Augmented.csv')
    return df

df = load_data()

# 3. Sidebar Filters
st.sidebar.image("https://upload.wikimedia.org/wikipedia/commons/f/f5/Roche_Logo.svg", width=150)
st.sidebar.title("Strategic Navigation")
st.sidebar.markdown("Filter insights dynamically to analyze specific local performance indicators.")

selected_product = st.sidebar.multiselect(
    "Filter by Roche Product:",
    options=df['Product_Clean'].unique(),
    default=df['Product_Clean'].unique()
)

selected_territory = st.sidebar.multiselect(
    "Filter by Therapeutic Area:",
    options=df['Territory'].unique(),
    default=df['Territory'].unique()
)

# Apply filters to the dataframe
filtered_df = df[
    (df['Product_Clean'].isin(selected_product)) &
    (df['Territory'].isin(selected_territory))
]

# 4. Dashboard Main Header
st.title("🇹🇳 Strategic Field Intelligence Platform")
st.subheader("Roche Tunisia Affiliate — Cross-Functional KPI Dashboard")
st.markdown("Transforming qualitative healthcare ecosystem data into automated operational actions.")

st.markdown("---")

# 5. KPI Highlight Cards (Top Row)
total_insights = len(filtered_df)
actionable_count = filtered_df['Is_Actionable'].sum()
actionable_rate = (actionable_count / total_insights * 100) if total_insights > 0 else 0

kpi1, kpi2, kpi3 = st.columns(3)
with kpi1:
    st.metric(label="Total Processed Insights", value=f"{total_insights} rows")
with kpi2:
    st.metric(label="Actionable Red Flags Identified", value=f"{actionable_count} insights")
with kpi3:
    st.metric(label="Field Actionability Rate", value=f"{actionable_rate:.1f}%")

st.markdown("---")

# 6. Graphs Grid Layout (Middle Row)
col1, col2 = st.columns(2)

with col1:
    st.subheader("📊 Volume Share of Strategic Pillars")
    # Sum up all the binary column counts
    target_cols = [
        'Cat_Access_And_Policy', 'Cat_Competitor_Intelligence',
        'Cat_Medical_Education_&_Practice', 'Cat_Operations_And_Diagnostics',
        'Cat_Stakeholder_Mapping', 'Cat_Strategic_Partnership', 'Cat_Promotional_Activity'
    ]

    category_counts = filtered_df[target_cols].sum().reset_index()
    category_counts.columns = ['Strategic Pillar', 'Insight Count']
    # Format labels for cleaner layout
    category_counts['Strategic Pillar'] = category_counts['Strategic Pillar'].str.replace('Cat_', '').str.replace('_', ' ')

    fig_pillar = px.bar(
        category_counts.sort_values(by='Insight Count', ascending=True),
        x='Insight Count',
        y='Strategic Pillar',
        orientation='h',
        color='Insight Count',
        color_continuous_scale='Blugrn',
        text_auto=True
    )
    fig_pillar.update_layout(showlegend=False, height=350)
    st.plotly_chart(fig_pillar, use_container_width=True)

with col2:
    st.subheader("🏥 Top Institutional Account Hotspots (HCOs)")
    # Get top 8 hospitals generating strategic data
    hco_counts = filtered_df['hco'].value_counts().head(8).reset_index()
    hco_counts.columns = ['Hospital / HCO', 'Total Insights']

    fig_hco = px.bar(
        hco_counts,
        x='Hospital / HCO',
        y='Total Insights',
        color='Total Insights',
        color_continuous_scale='Blues',
        text_auto=True
    )
    fig_hco.update_layout(xaxis_tickangle=-35, showlegend=False, height=350)
    st.plotly_chart(fig_hco, use_container_width=True)

st.markdown("---")

# 7. Lower Section: Product Matrix & Deep Dive Table
col3, col4 = st.columns([2, 3])

with col3:
    st.subheader("📦 Actionable Priority Matrix by Product")
    product_matrix = filtered_df.groupby('Product_Clean')['Is_Actionable'].sum().reset_index()
    product_matrix.columns = ['Product', 'Actionable Flares']

    fig_prod = px.pie(
        product_matrix,
        values='Actionable Flares',
        names='Product',
        hole=0.4,
        color_discrete_sequence=px.colors.sequential.Plotly3
    )
    fig_prod.update_layout(height=350)
    st.plotly_chart(fig_prod, use_container_width=True)

with col4:
    st.subheader("🔍 Real-time Insight Audit Stream")
    # Display raw details of selected records directly on screen
    display_df = filtered_df[['Date', 'Product_Clean', 'insight title', 'insight description', 'Category_ML']]
    st.dataframe(display_df, height=330, use_container_width=True)
```
